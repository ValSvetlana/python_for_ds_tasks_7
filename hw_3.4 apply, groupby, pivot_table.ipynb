 –ü–∏—Ç–∞–Ω–Ω—è –º–æ–∂—É—Ç—å –±—É—Ç–∏ –±—É–¥—å-—è–∫–æ—ó —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ, –∞–ª–µ —ó—Ö –º–∞—î –±—É—Ç–∏ 3. –ö–æ–∂–Ω–µ –ø–∏—Ç–∞–Ω–Ω—è –æ—Ü—ñ–Ω—é—î—Ç—å—Å—è —è–∫ 1 –∑–∞–≤–¥–∞–Ω–Ω—è. –ë–µ–∑ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è —Ü—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –î–ó –Ω–µ –ø—Ä–∏–π–º–∞—î—Ç—å—Å—è. –Ø–∫—â–æ —î –ø–∏—Ç–∞–Ω–Ω—è —â–æ–¥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è - –ø–∏—à—ñ—Ç—å —É —á–∞—Ç üôå


```python
#1. –Ø–∫–∞ —Å–µ—Ä–µ–¥–Ω—è —Å—É–º–∞ –ø–æ–∫—É–ø–∫–∏ –≤ –∫–æ–∂–Ω—ñ–π –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó —Ç–æ–≤–∞—Ä—ñ–≤ —É —Ä—ñ–∑–Ω–∏—Ö —Ñ—ñ–ª—ñ–∞–ª–∞—Ö? –º–æ–∂–µ —Å–ø–æ—Å—Ç—Ä–µ–≥—ñ–≥–∞—Ç–∏, —â–æ  —Ñ—ñ–ª—ñ–∞–ª–∏ —Å–ø–µ—Ü—ñ–∞–ª—ñ–∑—É—é—Ç—å—Å—è –Ω–∞ –ø—Ä–æ–¥–∞–∂—ñ –ø–µ–≤–Ω–∏—Ö –∫–∞—Ç–µ–≥–æ—Ä—ñ–π —Ç–æ–≤–∞—Ä—ñ–≤,
# —â–æ –ø—Ä–∏–∑–≤–æ–¥–∏—Ç—å –¥–æ –≤–∏—â–æ—ó —Å–µ—Ä–µ–¥–Ω—å–æ—ó —Å—É–º–∏ –ø–æ–∫—É–ø–∫–∏ —É —Ü–∏—Ö –∫–∞—Ç–µ–≥–æ—Ä—ñ—è—Ö.
# –û—Ä—ñ—î–Ω—Ç—É–≤–∞—Ç–∏ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤—ñ –∫–∞–º–ø–∞–Ω—ñ—ó –Ω–∞ —Ñ—ñ–ª—ñ–∞–ª–∏, –¥–µ —Å–µ—Ä–µ–¥–Ω—è —Å—É–º–∞ –ø–æ–∫—É–ø–∫–∏ –≤ –ø–µ–≤–Ω–∏—Ö –∫–∞—Ç–µ–≥–æ—Ä—ñ—è—Ö –Ω–∞–π–≤–∏—â–∞. –†–æ–∑–≥–ª—è–Ω—É—Ç–∏ –º–æ–∂–ª–∏–≤—ñ—Å—Ç—å —Ä–æ–∑—à–∏—Ä–µ–Ω–Ω—è –∞—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç—É —Ç–æ–≤–∞—Ä—ñ–≤ —É —Ü–∏—Ö –∫–∞—Ç–µ–≥–æ—Ä—ñ—è—Ö —É –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏—Ö —Ñ—ñ–ª—ñ–∞–ª–∞—Ö.
```
