3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = df.pivot_table(values='Invoice ID', index='Branch', columns='Product line', aggfunc='count', fill_value=0)
print(invoices_by_category)
```

```python
#1. Яка середня сума покупки в кожній категорії товарів у різних філіалах? може спострегігати, що  філіали спеціалізуються на продажі певних категорій товарів,
# що призводить до вищої середньої суми покупки у цих категоріях.
# Орієнтувати маркетингові кампанії на філіали, де середня сума покупки в певних категоріях найвища. Розглянути можливість розширення асортименту товарів у цих категоріях у відповідних філіалах.
```

```python
avg_category_branch = df.groupby(['Branch', 'Product line'])['Total'].mean().reset_index()
print(avg_category_branch)
```

```python
#2. Як впливає спосіб оплати (готівка, кредитна картка, дебетова картка) на прибуток у кожному з філіалів?
# які способи оплати є більш популярними і приносять більше прибутку у певних філіалах. Аналіз популярних способів оплати може допомогти визначити, які способи оплати варто додатково
# рекламувати або впроваджувати у філіалах для збільшення прибутку
```

```python
profit_by_payment_method_branch = df.groupby(['Branch', 'Payment'])['Total'].sum().reset_index()
print(profit_by_payment_method_branch)
```

```python
#3. Яка середня кількість товарів у чеку у різних філіалах? У філіалах з меншою кількістю товарів у чеку розробити стратегії зростання середнього чеку, наприклад,
# через акції "купуй більше - плати менше" або пропонування додаткових товарів на касі.
```

```python
avg_quantity_by_branch = df.groupby('Branch')['Quantity'].mean().reset_index()
print(avg_quantity_by_branch)
```
